Importing Libraries and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#We feed the data into a dataframe with the name 'dataset'
dataset = pd.read_csv('train.csv')
```

Data Exploration

```python
dataset.info()
#This is a quick summary of the content of the dataset along with the datatypes
```

```python
#Now, we get an analysis of the content of dataset, like the mean, standard deviation, quartiles of the data

dataset.describe()
```

```python
#Number of Numerical Columns in the Dataset

len(dataset.select_dtypes(include=['int64','float64']).columns)
```

```python
#Categorical Columns in the Dataset

print(dataset.select_dtypes(include=['object']).columns)

len(dataset.select_dtypes(include=['object']).columns)
```

Dealing with Null Values

```python
#Null values will cause our dataset to face irregularities
#We check for null values in the dataset using 'isnull' function

dataset.isnull().values.any()
```

```python
#We check the number of all the null values in the dataset

dataset.isnull().values.sum()
```

```python
dataset.isnull().sum()
#This gives us an idea of which columns have how many null values
```

```python
#Columns with Null values

dataset.columns[dataset.isnull().any()]
```

```python
len(dataset.columns[dataset.isnull().any()])
```

```python
#Null Values with Heatmap

plt.figure(figsize=(16,9))
sns.heatmap(dataset.isnull())
plt.show()
```

```python
dataset.shape
```

```python
null_percent = dataset.isnull().sum()/dataset.shape[0]*100
```

```python
plt.figure(figsize=(16,9))
null_percent.plot.bar()

#We use a barplot to visualize the percentage of null values in specific columns
```

```python
cols_to_drop = null_percent[null_percent > 50].keys()

#We create a list of the columns which have more than 50 percent null values in them
```

```python
cols_to_drop
```

```python
dataset = dataset.drop(columns=['Alley', 'PoolQC', 'Fence', 'MiscFeature'])

#We drop the columns which have more than 50 percent null values
```

```python
dataset.shape
```

```python
dataset.columns[dataset.isnull().any()]
```

```python
len(dataset.columns[dataset.isnull().any()])
```

```python
dataset['LotFrontage'] = dataset['LotFrontage'].fillna(dataset['LotFrontage'].mean())
dataset['MasVnrArea'] = dataset['MasVnrArea'].fillna(dataset['MasVnrArea'].mean())
dataset['GarageYrBlt'] = dataset['GarageYrBlt'].fillna(dataset['GarageYrBlt'].mean())

#We use the mean of these columns to replace the null values
```

```python
len(dataset.columns[dataset.isnull().any()])
```

```python
dataset.select_dtypes(include='object').columns
```

```python
dataset.columns[dataset.isnull().any()]
```

```python
len(dataset.columns[dataset.isnull().any()])
```

```python
dataset['MasVnrType'] = dataset['MasVnrType'].fillna(dataset['MasVnrType'].mode()[0])
dataset['BsmtQual'] = dataset['BsmtQual'].fillna(dataset['BsmtQual'].mode()[0])
dataset['BsmtCond'] = dataset['BsmtCond'].fillna(dataset['BsmtCond'].mode()[0])
dataset['BsmtExposure'] = dataset['BsmtExposure'].fillna(dataset['BsmtExposure'].mode()[0])
dataset['BsmtFinType1'] = dataset['BsmtFinType1'].fillna(dataset['BsmtFinType1'].mode()[0])
dataset['BsmtFinType2'] = dataset['BsmtFinType2'].fillna(dataset['BsmtFinType2'].mode()[0])
dataset['Electrical'] = dataset['Electrical'].fillna(dataset['Electrical'].mode()[0])
dataset['FireplaceQu'] = dataset['FireplaceQu'].fillna(dataset['FireplaceQu'].mode()[0])
dataset['GarageType'] = dataset['GarageType'].fillna(dataset['GarageType'].mode()[0])
dataset['GarageFinish'] = dataset['GarageFinish'].fillna(dataset['GarageFinish'].mode()[0])
dataset['GarageQual'] = dataset['GarageQual'].fillna(dataset['GarageQual'].mode()[0])
dataset['GarageCond'] = dataset['GarageCond'].fillna(dataset['GarageCond'].mode()[0])

#We use the mode of these columns to replace the null values
```

```python
len(dataset.columns[dataset.isnull().any()])
```

```python
dataset.isnull().values.any()
```

Distplot

```python
plt.figure(figsize=(16,9))
bar = sns.distplot(dataset['SalePrice'])
bar.legend(['Skewness: {:2f}'.format(dataset['SalePrice'].skew())])
plt.show()

#This displot allows us to identify the skewness and visualize the average price range of the sale price
```

Correlation Matrix

```python
dataset_2 = dataset.drop(columns='SalePrice')
```

```python
dataset_2.shape
```

```python
dataset_2.corrwith(dataset['SalePrice']).plot.bar(
    figsize=(16,9), title='Correlation with Sale Price', grid = True
)

#We build a visualisation of the correalation of different columns with the Sale Price
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(dataset.corr(), cmap='coolwarm', annot=True, linewidths=2)

#We build a correlation map to identify the most correlated columns to the sale price
```

```python
high_corr = dataset.corr()
```

```python
high_corr_features = high_corr.index[abs(high_corr['SalePrice']) > 0.5]

# We identify the columns which have a correlation with the Sale Price greater than 50 percent
```

```python
high_corr_features
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(dataset[high_corr_features].corr(), cmap='coolwarm', annot=True, linewidths=2)

# We plot a heatmap of the highly correlated columns with the Sale Price
```

```python
fig, ((ax1,ax2), (ax3,ax4), (ax5, ax6)) = plt.subplots(nrows = 3, ncols = 2, figsize=(14,10))

sns.regplot(x = 'OverallQual', y = 'SalePrice', data=dataset,scatter=True, fit_reg=True, ax = ax1)

sns.regplot(x = 'GrLivArea', y = 'SalePrice', data=dataset,scatter=True, fit_reg=True, ax = ax2)

sns.regplot(x = 'GarageArea', y = 'SalePrice', data=dataset,scatter=True, fit_reg=True, ax = ax3)

sns.regplot(x = 'FullBath', y = 'SalePrice', data=dataset,scatter=True, fit_reg=True, ax = ax4)

sns.regplot(x = 'TotalBsmtSF', y = 'SalePrice', data=dataset,scatter=True, fit_reg=True, ax = ax5)

sns.regplot(x = 'YearBuilt', y = 'SalePrice', data=dataset,scatter=True, fit_reg=True, ax = ax6)

# This is a visualisation of the most highly correlated columns to the sale price
```
